```python
import pandas as pd
import numpy as np
```

```python
#imports xT grid
xT = pd.DataFrame(pd.read_excel("xT map.xlsx", header = None))
xTmap = np.array(xT)
```

```python
#this section imports data and filters out any rows that are not successful passes
DF = pd.read_csv('sancho_actions_merged.csv')
DF = DF[['Event ID', 'Player', 'Minute','Second','X', 'Y', 'Type', 'Outcome','endX', 'endY', 'isAssist', 'Set Piece',
        'Left', 'Center', 'Right']]
DF = DF[DF['Type']=='Pass']
DF = DF[DF['Set Piece']=='Blank']
xThreat = DF[DF['Outcome']=='Pos']
xThreat.head()
```

```python
# in this part, we assigne each row an xBin and a yBin value
# Bin values correspond to boxes in the xT grid

yb = 100/17
xb = 100/21

i = 0
j = 0
e = 0
k = 0

x1_bin = []
y1_bin =[]

x2_bin = []
y2_bin = []

for e in range(len(xThreat['Y'])):
    i = xThreat['Y'].iloc[e]
    y1 = int((i//yb))
    j = xThreat['X'].iloc[e]
    x1 = int((j//xb))
    
    y1_bin.append(y1)
    x1_bin.append(x1)
    
xThreat['x1_bin'] = x1_bin
xThreat['y1_bin'] = y1_bin

for e in range(len(xThreat['Y'])):
    i = xThreat['endY'].iloc[e]
    y2 = int((i//yb))
    j = xThreat['endX'].iloc[e]
    x2 = int((j//xb))
    
    y2_bin.append(y2)
    x2_bin.append(x2)
    
xThreat['x2_bin'] = x2_bin
xThreat['y2_bin'] = y2_bin    

xThreat
```

```python
'''
After we have assigned each row bin values, we get the corresponding starting and ending xT values by comapring bin
values to the xT grid
'''
starting_xT = []
ending_xT = []
xT_change = []
converter = {16:0, 15:1, 14:2, 13:3, 12:4, 11:5, 10:6, 9:7, 8:8, 7:9, 6:10, 5:11, 4:12, 3:13, 2:14, 1:15, 0:16}

for element in range(len(x1_bin)):
    y1 = y1_bin[element]
    y1 = converter[y1]
    x1 = x1_bin[element]
    
    starting_xT.append(xTmap[y1][x1]*100)
    
for element in range(len(x2_bin)):
    y2 = y2_bin[element]
    y2 = converter[y2]
    x2 = x2_bin[element]
    
    ending_xT.append(xTmap[y2][x2]*100)
    
for element in range(len(starting_xT)):
    value = ending_xT[element] - starting_xT[element]
    xT_change.append(value)
    

xThreat['starting_xT'] = starting_xT
xThreat['ending_xT'] = ending_xT
xThreat['xT_change'] = xT_change

xThreat
```

```python
xT_net = xThreat['xT_change'].sum()
xT_net
```

```python
total_90s = 386/90
total_90s
```

```python
passes_from_left = DF.Left.value_counts()['Yes']
passes_from_center = DF.Center.value_counts()['Yes']
passes_from_right = DF.Right.value_counts()['Yes']

print(passes_from_left)
print(passes_from_center)
print(passes_from_right)
```

```python
#Net sum of xT from the left, right and center
Left = xThreat[xThreat['Left']=='Yes']
Left_Total = Left['xT_change'].sum()

Center = xThreat[xThreat['Center']=='Yes']
Center_Total = Center['xT_change'].sum()

Right = xThreat[xThreat['Right']=='Yes']
Right_Total = Right['xT_change'].sum()


#xT per touch from the left, right and center
xTpP_Left = Left_Total/passes_from_left
xTpP_Center = Center_Total/passes_from_center
xTpP_Right = Right_Total/passes_from_right
xTpP = xT_net/232

print('Left xT =', Left_Total)
print('Center xT =', Center_Total)
print('Right xT =', Right_Total)
print('\n')
print('xT per 90 =', xT_net/total_90s) 
print('\n')
print('Left xT per pass =', xTpP_Left)
print('Center xT per pass =', xTpP_Center)
print('Right xT per pass =', xTpP_Right)
print('xT per pass =', xTpP)
```
